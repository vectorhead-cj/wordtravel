## ENABLE word list

https://www.freescrabbledictionary.com/enable/

Lemmas + conjugations, no frequency data

```python
import os

with open("enable_20251129.txt", "r", encoding="utf-8") as f:
    enable_words = [line.strip() for line in f if line.strip()]

output_dir = os.path.join("..", "WordTravel", "src", "data")
os.makedirs(output_dir, exist_ok=True)

for word_length in [3, 4, 5, 6]:
    filtered_enable_words = []
    for word in enable_words:
        if not word.isalpha() or not word.islower():
            continue
        if len(word) == word_length:
            filtered_enable_words.append(word)

    filename = os.path.join(output_dir, f"words_en_{word_length}.txt")
    with open(filename, "w", encoding="utf-8") as f:
        for word in filtered_enable_words:
            f.write(f"{word}\n")
    print(f"Length {word_length}: wrote {len(filtered_enable_words)} words to {filename}")
    
```

## ENABLE sorted by Wordfreq

Use ENABLE as the base, but "left join" frequencies from Wordfreq. Sort by highest frequency, nulls at the bottom.

```python
#!pip install tqdm
```

```python
import os
from tqdm import tqdm
from wordfreq import word_frequency

with open("enable_20251129.txt", "r", encoding="utf-8") as f:
    enable_words = [line.strip() for line in f if line.strip()]

output_dir = os.path.join("..", "WordTravel", "src", "data")
os.makedirs(output_dir, exist_ok=True)

for word_length in tqdm([3, 4, 5, 6]):
    filtered = [
        w for w in enable_words
        if len(w) == word_length and w.isalpha() and w.islower()
    ]
    word_freqs = [(w, word_frequency(w, "en", wordlist="large")) for w in filtered]
    word_freqs.sort(key=lambda x: -x[1])

    filename = os.path.join(output_dir, f"words_en_{word_length}.txt")
    with open(filename, "w", encoding="utf-8") as f:
        for word, freq in word_freqs:
            f.write(f"{word}\t{freq:.6e}\n")
    print(f"Length {word_length}: wrote {len(word_freqs)} words to {filename}")
```

```python
from wordfreq import top_n_list
from wordfreq import word_frequency
import re
```

### Svenska

```python
words_se = top_n_list('sv', 10000, wordlist='small')

for word in words_se:
	if any(c.isupper() for c in word):
		print(word)

pattern_se = re.compile(r"^[a-zåäö]+$", re.IGNORECASE)
filtered_se = [w for w in words_se if pattern_se.match(w)]
filtered_se = list(dict.fromkeys(filtered_se))

for word_length in [2, 3, 4, 5, 6, 7, 8, 9]:
    filtered_se_length = [w for w in filtered_se if len(w) == word_length]
    filename = f'words_se_{word_length}.txt'
    with open(filename, 'w', encoding='utf-8') as f:
        for word in filtered_se_length:
            f.write(f"{word}\n")
```
